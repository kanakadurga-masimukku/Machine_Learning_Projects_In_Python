# Calculating ADFULLER TEST COMPONENTS

```python
y=pd.read_csv("C:\\Users\\KATA\\Documents\\DATA SCIENCE\\DataSets 1\\df_vacation.csv",header=0,index_col=0,squeeze=True)
```

```python
z=y.values
```

```python
c=adfuller(z)
```

```python
print("ADF STATIC:",c[0])
```

```python
print("P VALUE:",c[1])
```

```python
print("CRITICAL VALUES:",c[4])
```

# Plotting ACF And PACF Plots

```python
plot_acf(y,lags=40)
plt.show()
```

```python
plot_pacf(y,lags=20)
plt.show()
```

# ForeCasting The Time Series Using Arima Model

```python
a=ARIMA(y,order=(1,1,1))
```

```python
b=a.fit()
```

```python
print(b.summary())
```

```python
b.plot_predict()
plt.legend()
```
